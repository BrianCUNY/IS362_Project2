# Kaldheim card search

### I have created a .CSV for my current MTG card collection that was scanned in another app.
### I am looking to find all of the cards I have scanned, in the newest Kaldheim set/expansion.

```python
# Location of my exported card data in .CSV format 
data = pd.read_csv(r'C:\Users\brian\Downloads/TCGplayer.csv')
```

```python
# Creating a DF for imported data, specifying column structure within .CSV
df = pd.DataFrame(
    data,
    columns=[
        'Quantity',
        'Name',
        'Simple Name',
        'Set',
        'Card Number',
        'Set Code',
        'Printing',
        'Condition',
        'Language',
        'Rarity',
        'Product ID',
        'SKU'
    ]
)
```

```python
print(df)
```

```python
# Cleaning up the data, I don't need the following columns that are either duplicates or irrelevant. 
df.drop(columns=[
    'Simple Name', 
    'Card Number', 
    'Printing', 
    'Condition', 
    'Language', 
    'Product ID', 
    'SKU'
]
       )
```

```python
# Filtering data to provide only cards belonging to the Kaldheim set.
# Another option would be to utilize 'Set Code' for 'KHM' which is the Kaldheim set.
kaldheimcards = df[df['Set']=='Kaldheim']
```

```python
# Here is the printout of all Kaldheim cardsin my collection.
print(kaldheimcards)
```
